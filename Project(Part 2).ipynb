```python
gen_type = []
gen = []
for i in d['Processors']:
    x=str(i).split('(')
    gen.append(x)
for i in gen:
    if(len(i) == 2):
        if(i[1][0] == '1'):
            gen_type.append('10')
        else:
            gen_type.append(i[1][0])
    else:
        gen_type.append('Not Mentioned')
d['Gen_type'] = gen_type
d['Gen_type']
```

```python
ram_in_gb = []
for i in d['RAM']:
    if(i[0] == '8'):
        ram_in_gb.append(i[0])
    elif(i[0] == '4'):
        ram_in_gb.append(i[0])
    elif(i[0] == '1'): 
        ram_in_gb.append('16')
    elif(i[0] == '3'):
        ram_in_gb.append('32')
    elif(i[0]=='2'):
            if(i[1]=='4'):
                ram_in_gb.append('24')
            else:
                ram_in_gb.append('2')
    else:
        ram_in_gb.append('Null')
print(len(ram_in_gb))
d['Ram_in_gb'] = ram_in_gb
d['Ram_in_gb']
```

```python
d
```

```python
disk_drive = []
x=[]
for i in d['Storage']:
    x.append(str(i))
for i in x:
    if('HDD' in i and 'SSD' not in i):
        disk_drive.append('HDD')
    elif('SSD' in i and 'HDD' not in i):
        disk_drive.append('SSD')
    elif('HDD' in i and 'SSD' in i):  
        disk_drive.append('Both')
    else:
        disk_drive.append('Not Metioned')
d['Disk_drive'] = disk_drive
d['Disk_drive']
```

```python
d
```

```python
processor_name = []
x=[]
for i in d['Processors']:
    x.append(str(i))
for i in x:
    if('Intel' in i):
        processor_name.append('Intel')
    elif('AMD' in i):
        processor_name.append('AMD')
    elif('Microsoft' in i):
        processor_name.append('Microsoft')
    else:
        processor_name.append('Not Metioned')
d['Processor_Name'] = processor_name
d['Processor_Name']
```

```python
d['processors'] = d['Processors'].copy() # Creating a copy of Processor
d['processors'].replace(np.nan,'Not metioned',inplace=True)
processor_type = []
z=[]
for i in d['processors']:
    z.append(str(i))
for i in z:
    if(i[11] == 'i' or i[11] == 'm'):
        processor_type.append(i[11:13])
    elif(i[10] == '5'):
        processor_type.append('Ryzen 5')
    elif(i[10]  == '3'):
        processor_type.append('Ryzen 3')
    elif(i[10] == '7'):
        processor_type.append('Ryzen 7')
    elif('Celeron' in i):
        processor_type.append('Celeron')
    elif('SQ1' in i):
        processor_type.append('SQ1')
    elif('Pentium' in i):
        processor_type.append('Pentium')
    elif('APU' in i and 'Athlon' not in i):
        processor_type.append('APU')
    elif('Athlon' in i and 'APU' not in i):
        processor_type.append('Athlon')
    elif('APU' in i and 'Athlon' in i):
        processor_type.append('APU and Athlon')
    else:
        processor_type.append('Not Metioned')
d['processor_type'] = processor_type
d['processor_type']
```

```python
os = []
for i in d['OS']:
  if('Windows 10' in i): # Windows Operating System
    os.append('Windows 10')
  elif('Windows 8' in i): # Windows Operating System
        os.append('Windows 8')
  elif('Mac' in i):   # Mac Operating System
    os.append('Mac')
  elif('Chrome' in i):   # Mac Operating System
    os.append('Mac')
  elif('Linux' in i): # Linux Operating System
    os.append('Linux/Ubuntu')
  elif('DOS' in i):   # DOS Operating System
    os.append('DOS')
d['Operating_System'] = os
d['Operating_System']
```
